# **WSN Project** - Central Hidrelétrica Cachoeira Dourada S.A.

### **Wireless Sensor Networks**
A wireless sensor network (WSN) is a set of transducers with a communication infrastructure used to monitor physical or environmental conditions, such as temperature, sound, pressure, speed, illumination intensity, sound intensity, chemical concentrations, pollutant levels, Body functions etc. In this case, the data comes from the field of the Cachoeira Dourada Hydroelectric Plant, availables at this link: http://cloud.traceback.com.br/wsn/wsn_001/journal_ufsc.html.

The state of the system can be seen in snapshots at this link: http://cloud.traceback.com.br/wsn/dashlist_cdsa.html. It is important to note that the system has been live since 2017-09-20, with no reset. The link drops that occurred in this period were recovered by the Zigbee PRO stack. The most stable period was from July / 2018, before that we noticed many problems. The UG3 and UG5 routers are powered by AC from the dam, so each time they do maintenance, these nodes fall off and take some others along.And the router diagram can be seen in the coordinadores page

The systems consist in 7 Routers and 1 Coordinator, total of 8 Routers. Taking for example the information related to 2018/01/01 we can extract the following conclusions. 
1. There are 8 locations and 1 Router for every location

2. Each router gives information about it's tempreature and bus voltage using channels 7 and 6 respectively and a timestamp.

3. There are other sensors in addition of the routers. Each sensor communicates with one specific router and a Router can communicate with multiple sensors

4. The information about the dim is given by these sensors

5. The **Radios** communicates with some modules, these modules provides other informations. Every Radio has some modules.

6. In <span style="color:red">Montante UG5</span> location is Router <span style="color:blue">57.FE.0E</span>, wich communicates with Modules: 
    * Acquisition - Temperature **E5.5A.24**  that extract info from (2) PT100 using **channels 0 and 1**, and it's temperature and VBus using channel **7 and 6** 
    * Acquisition - Current / Voltage **CB.0A.C0** that extract info about the level in the dim using **channel 1**, and it's temperature and VBus using channel **7 and 6**
    * Power - Solar Panel **B2.9F.A9** that extract the MPPT voltage and the voltage from the solar panel using **channels 0 and 1**.


7. In <span style="color:red">Poço de Drenagem A</span> location is Router <span style="color:blue">57.FE.0F</span>, wich communicates with: 
    * Acquisition - Current / Voltage **E9.39.32**  which is a level probe , uses **channel 1**.
    * Power - AC/DC Input **0D.82.38** that measures voltage from Battery and Supply voltage using **channel 0 and 1**, and it's temperature and VBus using channel **7 and 6**


8. In <span style="color:red">Transformador Elevador</span> location is Router <span style="color:blue">57.FE.03</span>, wich communicates with: 
    * Acquisition - Temperature sensor **8D.AC.91**  which gives 2 temperatures (Oil and Gabinetes) through **channels 0 and 1**.
    * Power - Solar Panel sensor **39.E2.80** that extract the MPPT voltage and the voltage from the solar panel using **channels 0 and 1**


9. The <span style="color:red">Montante UG5 and Transformador Elevador Routers</span> are powered by Solar energy.

10. The <span style="color:red">Poço de Drenagem A Router</span> is powered by AC energy from the Hydroelectric plant.

```python
def join_feature(df1, df2, join_create = 'create', module='29.E5.5A.24', channels=['PT100(0)','PT100(1)']):
    '''
    This function returns (1) a Datafram indexed from 0 to n
    formed by the entries of the df1 that matches
    with module and joints it with channels 0 and 1 of df2
    (2 features of df2), this dataframe is a slice of
    the original dataset. 
    Or (2) The original dataset joined with the values of 
    channel 0 and 1 for a specific module. The resulting 
    dataset will have NaN values in the modules that doesn't
    match
    '''
    if (len(df1.columns) == 7) & (module=='29.E5.5A.24'):
        df1 = df1.drop(['PT100(0)','PT100(1)'], axis=1)
    if (len(df1.columns) == 8) & (module=='A7.CB.0A.C0'):
        df1 = df1.drop(['LVL_Dim(1)'], axis=1)
    if (len(df1.columns) == 10) & (module=='34.B2.9F.A9'):
        df1 = df1.drop(['V_MPPT', 'V_Panel'], axis=1)
    
    df_temporal = df1[df1['Module'] == module]
    
    if join_create == 'create':
        df_temporal = df_temporal.reset_index(inplace=False)
        df_temporal = df_temporal.drop('index', axis=1)
        return pd.concat([df_temporal,df2[channels]], axis=1)
    else:
        df2_temp = df2.set_index(df_temporal.index)
        dftempp = pd.concat([df_temporal, df2_temp[channels]],axis=1)
        return pd.concat([df1, dftempp[channels]], axis=1)

#df = join_feature(df, df_pt100,'join')
#df_sliced = join_feature(df, df_pt100,'create')
#df = join_feature(df, df_dim,'join', module='A7.CB.0A.C0', channels=['LVL_Dim(1)'])
#df_sliced = join_feature(df, df_pt100,'create')
#df = join_feature(df, df_vpanel,'join', module='34.B2.9F.A9', channels=['V_MPPT', 'V_Panel'])
#df_sliced = join_feature(df, df_vpanel,'create',module='34.B2.9F.A9', channels=['V_MPPT', 'V_Panel'])
#df.describe().transpose()
```
